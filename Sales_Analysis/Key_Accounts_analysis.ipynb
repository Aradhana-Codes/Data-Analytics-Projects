```python
import numpy as np
import re

def convert_boq(value):
    # Handle missing values
    if pd.isna(value):
        return np.nan

    # If already numeric, return as float
    if isinstance(value, (int, float)):
        return float(value)

    # Convert everything else to string
    value = str(value).lower().strip()

    # Extract numbers
    numbers = re.findall(r'\d+', value)

    if len(numbers) == 1:
        return float(numbers[0])

    elif len(numbers) == 2:
        return (float(numbers[0]) + float(numbers[1])) / 2

    else:
        return np.nan

# On average, this account is expected to generate 1.5 BOQs per year
```

```python
data['estimated_boq_clean'] = data['estimated_vol_of_boq_per_year_with_ejc'].apply(convert_boq)
```
