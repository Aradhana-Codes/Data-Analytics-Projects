```python
data=pd.read_excel("Sample Month End Report for Assesment.xlsx",sheet_name="Inquiry Tracker")
data
```

```python
data=data.drop(columns=['Unnamed: 16','Unnamed: 17','Unnamed: 18','Unnamed: 19','Unnamed: 20','Unnamed: 21'])
```

```python
data=data.drop(columns=['Unnamed: 6'])
```

```python
data=data.drop(columns=['Feasability Ratio'])
```

```python
data.isnull().sum()
```

```python
data = data.drop(index=range(26, 30))
```

```python
data=data.drop(columns=['Lost Reason'])
```

```python
data.isnull().sum()
```

```python
data
```

```python
data = data.drop(index=range(1))
```

```python
data
```

```python
data['Applicator'].fillna('Not Assigned', inplace=True)
```

```python
data['Number of Hours'].fillna(0, inplace=True)
data['Value/Hours'].fillna(0, inplace=True)
data['Quantity in LM'].fillna(0, inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Inquiry Date'].fillna(method='ffill', inplace=True)
```

```python
data.isnull().sum()
```

```python
data
```

```python
data['Applicator'] = data['Applicator'].replace({
    'No': 'Not Assigned',
    'no': 'Not Assigned',
    'Not assigned': 'Not Assigned'
})
```

```python
data['Applicator'].unique()
```

```python
data['Applicator'] = data['Applicator'].replace({
    'Asg win': 'Asg Win',    
})
```

```python
data['Applicator'].unique()
```

```python
data['Projects'].unique()
```

```python
data['month'] = data['Inquiry Date'].dt.to_period('M')
monthly_hours = data.groupby('month')['Projects'].count().sort_values(ascending=False)
monthly_hours
```

# Project inquiries show a clear seasonal pattern, 
# peaking in April 2025 with sustained activity across mid-2025, 
# indicating Q2 as the strongest period for new business.
