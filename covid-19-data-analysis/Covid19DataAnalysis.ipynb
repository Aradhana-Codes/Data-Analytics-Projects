# Data Visualization

# Univariate Analysis

```python
# Set up the figure
plt.figure(figsize=(18,5))
# Histogram for Confirmed cases
plt.subplot(1, 3, 1)  # 1 row, 3 columns, 1st plot
sns.histplot(data['Confirmed'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Confirmed Cases')
plt.xlabel('Confirmed Cases')
plt.ylabel('Frequency')

# Histogram for Deaths
plt.subplot(1, 3, 2)  # 1 row, 3 columns, 2nd plot
sns.histplot(data['Deaths'], bins=30, kde=True, color='salmon')
plt.title('Distribution of Deaths')
plt.xlabel('Deaths')
plt.ylabel('Frequency')

# Histogram for Recovered
plt.subplot(1, 3, 3)  # 1 row, 3 columns, 3rd plot
sns.histplot(data['Recovered'], bins=30, kde=True, color='lightgreen')
plt.title('Distribution of Recovered Cases')
plt.xlabel('Recovered Cases')
plt.ylabel('Frequency')

# Show all plots together
plt.tight_layout()
plt.show()
```

```python
# Confirmed: Mostly low; few countries with very high cases (right-skewed).
# Deaths: Similar trend but smaller numbers.
# Recovered: Follows confirmed trend; varies by healthcare quality.
# COVID-19 impact was uneven across countries — few had very high cases and deaths, while recovery rates differed due to healthcare differences.
```

# Bivariate Analysis

```python
# Scatter Plot — Confirmed vs Deaths
plt.figure(figsize=(8,6))
sns.scatterplot(x='Confirmed', y='Deaths', data=data, color='purple')
plt.title('Confirmed Cases vs Deaths')
plt.xlabel('Confirmed')
plt.ylabel('Deaths')
plt.show()
```

```python
# A positive correlation exists — more confirmed cases generally lead to more deaths.
# Countries with higher infection rates also recorded higher deaths, indicating severity correlation
```

```python
# Scatter Plot — Confirmed vs Recovered
plt.figure(figsize=(8,6))
sns.scatterplot(x='Confirmed', y='Recovered', data=data, color='green')
plt.title('Confirmed Cases vs Recovered')
plt.xlabel('Confirmed')
plt.ylabel('Recovered')
plt.show()
```

```python
# This indicates that as confirmed cases rise in a country, the number of recovered cases also proportionally increases, as expected over time. 
# The presence of a few extreme data points far from the origin highlights the handful of countries that account for the vast majority of global COVID-19 cases and recoveries.
```

```python
# Regression Plot — Deaths vs Recovered
plt.figure(figsize=(8,6))
sns.regplot(x='Recovered', y='Deaths', data=data, color='orange')
plt.title('Deaths vs Recovered with Regression Line')
plt.xlabel('Recovered')
plt.ylabel('Deaths')
plt.show()
```

```python
# As recoveries increase, deaths also increase slightly but not proportionally.
# Recovery and death both depend on the total cases, but recovery dominates (more recovered than deaths).
```

# Categorical Visualization

```python
#Bar Plot — Top 10 Countries by Confirmed Cases
top10 = data.sort_values('Confirmed', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='Confirmed', y='Country/Region', data=top10, palette='viridis')
plt.title('Top 10 Countries by Confirmed Cases')
plt.xlabel('Confirmed Cases')
plt.ylabel('Country')
plt.show()
```

```python
# Shows countries with the highest death rate per 100 cases.
# These countries faced serious healthcare challenges and delayed pandemic control.
```

```python
# Pie Chart — Deaths Share in Top 5 Countries
top5 = data.sort_values('Deaths', ascending=False).head(5)
plt.figure(figsize=(6,6))
plt.pie(top5['Deaths'], labels=top5['Country/Region'], autopct='%1.1f%%', startangle=90)
plt.title('Share of Deaths Among Top 5 Countries')
plt.show()
```

```python
# US>Brazil>United Kingdom>Mexico>Italy 
```

```python
# Group data by WHO Region and sum the cases
region_data = data.groupby('WHO Region')[['Confirmed','Deaths','Recovered']].sum().reset_index()
plt.figure(figsize=(15,5))

# 1. Confirmed cases
plt.subplot(1,3,1)
sns.barplot(x='WHO Region', y='Confirmed', data=region_data, palette='Blues_r')
plt.title('Confirmed Cases by WHO Region')
plt.xticks(rotation=45)
plt.ylabel('Confirmed Cases')

# 2. Deaths
plt.subplot(1,3,2)
sns.barplot(x='WHO Region', y='Deaths', data=region_data, palette='Reds_r')
plt.title('Deaths by WHO Region')
plt.xticks(rotation=45)
plt.ylabel('Deaths')

# 3. Recovered
plt.subplot(1,3,3)
sns.barplot(x='WHO Region', y='Recovered', data=region_data, palette='Greens_r')
plt.title('Recovered Cases by WHO Region')
plt.xticks(rotation=45)
plt.ylabel('Recovered Cases')

plt.tight_layout()
plt.show()
```

```python
# Eastern Mediterranean or Americas might have higher confirmed,death and recovered counts.    
```

```python
# Pie Chart (Proportion of Confirmed Cases by Region)
plt.figure(figsize=(8,8))
plt.pie(region_data['Confirmed'], labels=region_data['WHO Region'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Proportion of Confirmed Cases by WHO Region')
plt.show()
```

```python
# Sort data by '1 week change' in descending order and select top 10
top10_new_cases = data.sort_values(by='1 week change', ascending=False).head(10)

# Display top 10 countries
top10_new_cases[['Country/Region','1 week change']]
plt.figure(figsize=(12,6))
sns.barplot(x='1 week change', y='Country/Region', data=top10_new_cases, palette='Reds_r')
plt.title('Top 10 Countries by New Infections in Last Week')
plt.xlabel('New Cases in Last Week')
plt.ylabel('Country')
plt.show()
```

```python
# the data clearly indicates a severe and disproportionate burden of new infections in the United States, 
# followed by high rates in India and Brazil, with a much lower, more gradual distribution among 
# the other seven countries in the top 10
```

```python
# Select top 10 countries by New Cases
top10_cases = data.sort_values(by='New cases', ascending=False).head(10)
top10_cases = top10_cases.sort_values(by='New cases', ascending=True)  # for horizontal bars

# Select top 10 countries by New Deaths
top10_deaths = data.sort_values(by='New deaths', ascending=False).head(10)
top10_deaths = top10_deaths.sort_values(by='New deaths', ascending=True)

# Select top 10 countries by New Recovered
top10_recovered = data.sort_values(by='New recovered', ascending=False).head(10)
top10_recovered = top10_recovered.sort_values(by='New recovered', ascending=True)

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(20,8))

# Plot New Cases
sns.barplot(x='New cases', y='Country/Region', data=top10_cases, palette='Reds_r', ax=axes[0])
axes[0].set_title('Top 10 Countries by New Cases')
axes[0].set_xlabel('New Cases')
axes[0].set_ylabel('Country')

# Plot New Deaths
sns.barplot(x='New deaths', y='Country/Region', data=top10_deaths, palette='Greys_r', ax=axes[1])
axes[1].set_title('Top 10 Countries by New Deaths')
axes[1].set_xlabel('New Deaths')
axes[1].set_ylabel('')

# Plot New Recovered
sns.barplot(x='New recovered', y='Country/Region', data=top10_recovered, palette='Greens_r', ax=axes[2])
axes[2].set_title('Top 10 Countries by New Recovered')
axes[2].set_xlabel('New Recovered')
axes[2].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
# US has more new cases and more Deaths but Brazil has more New Recovered
```
